```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(inputs):
    sig_scores = [1/float(1+np.exp(-x)) for x in inputs]
    return sig_scores
```

```python
inputs = [2, 5, 3, 7]
print(sigmoid(inputs))
```

```python
def lineGraph(x, y, xTitle, yTitle):
    plt.plot(x, y)
    plt.xlabel(xTitle)
    plt.ylabel(yTitle)
    plt.show()
```

```python
graph_x = range(-10, 10)
graph_y = sigmoid(graph_x)
print("inputs:", graph_x)
print("outputs:", graph_y)
```

```python
lineGraph(graph_x, graph_y, "inputs", "sig_scores")
```

```python
def softmax(inputs):
    val = np.exp(inputs)/np.sum(np.exp(inputs))
    return val
```

```python
graph_x = range(-10, 10)
graph_y = softmax(graph_x)
print("inputs:", graph_x)
print("outputs:", graph_y)
```

```python
lineGraph(graph_x, graph_y, "inputs", "softmax_score")
```
