```python
import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np
```

```python
n = 1024 #number of nuerons
size = 500 #no.of tests (input examples)
binwidth = 5 #histogram bin width for data visualization
```

```python
for p in range(1, 10):
    prob = p/10.0 # per layer probability
    #print(prob)
    rnd_values = binom.rvs(n, prob, size=size) # generate size values from a bi(n, prob)
    #print(rnd_values)
    #print(np.random.rand(3, 1))
    plt.hist(rnd_values, bins=[x for x in range(0, n, binwidth)], normed=1, label=str(prob))
    
plt.legend(loc='upper right')
plt.show()
```
